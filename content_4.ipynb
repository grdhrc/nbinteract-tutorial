# Client Server Architecture

-  Client Server architecture is a network architecture in which the multiple clients request and receive service from a centralized server.
-  In client server architecture, none of the peers aids in distributing the file.
-  The server must transmit one copy of the file to each of the N peers.
-  The Main objective of this project is to compute the File Distribution times of Client Server , Peer to Peer architecture and   demonstrate the differences of respective distribution times.


# Minimum File Distribution Time for Client Server Architechture is given by

   ## $$Dcs=max\{\frac{NF}{_s},\frac{F}{dmin}\}$$
   

### The Parameters involved in computing  File Distribution time for Client server architecture are

- s= server upload rate
- dmin = is the download rate of peer with lowest download rate.
- N = Number of peers involved in transferring the file over the network.
- F = Total size of the file being transmitted.

# Peer to Peer Architecture

- ### In P2P architecture each peer can assist the server in distributing the file. 
- ### When a peer receives some file data, it can use its own upload capacity to redistribute the data to other peers.
- ### The total upload capacity of the system as a whole is equal to the upload rate of the server plus the upload rates of each of the individual peers, that is, utotal = us + u1 +  + uN.
- ### The system must deliver (upload) F bits to each of the N peers, thus delivering a total of NF bits. This cannot be done at a rate faster than utotal.


### The Parameters involved in computing  File Distribution time for Peer to Peer  architecture.

   ## $$Dp2p=max\{\frac{N*F}{_s},\frac{F}{dmin},\frac{N*F}{_s+\sum_{i=1}^{N} _i}\}$$

- s= server upload rate
- dmin = is the download rate of peer with lowest download rate.
- N = Number of peers involved in transferring the file over the network.
- F = Total size of the file being transmitted.
- Utotal=Total upload capacity of the system as a whole is equal to the upload rate of the server plus the upload rates of each of the individual peers.


# Interactive Graphs to demonstrate Client Server and Peer to Peer File Distribution times

```python
# nbi:hide_in
import ipywidgets as widgets
from ipywidgets import interact, interactive, fixed, interact_manual
from ipywidgets.embed import embed_minimal_html
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import mplcursors
%matplotlib inline

# nbi:hide_in
def client_server(n,F,us,dmin):
    return max(n*F/us, F/dmin)

def peer_to_peer(n,u,F,us,dmin):
    return max(F/us, F/dmin, n*F/(us + sum(u)))

def cliser(F,us,dmin):
    N = [10, 100, 200, 400, 600, 800, 1000]
    cs = [client_server(n,F,us,dmin) for n in N]
    plt.plot(N, cs, 'bo')
    plt.plot(N, cs, 'b', label='Client-Server')
    plt.title('Minimum Distribution Time for Client-Server Architecture')
    plt.xlabel('Number of Peers')
    plt.ylabel('Minimum Distribution Time (s)')
    mplcursors.cursor(hover=True)
    plt.show()
    return cs

def p2ps(F,us,dmin):
    N = [10, 100, 200, 400, 600, 800, 1000]
    u = [100, 200, 300, 400, 600, 800, 1000]
    p2p = [peer_to_peer(N[i], [u[i]] * N[i],F,us,dmin) for i in range(len(N))]
    plt.plot(N, p2p, 'ro')
    plt.plot(N, p2p, 'r', label='P2P')
    plt.title('Minimum Distribution Time for peer to peer Architecture')
    plt.xlabel('Number of Peers')
    plt.ylabel('Minimum Distribution Time (s)')
    mplcursors.cursor(hover=True)
    plt.show()
    return p2p

# nbi:hide_in
y=interactive(cliser,F=(10**4,10**7,1000),us=(1000,20000,100),
              dmin=(10,2000,10))
x=interactive(p2ps,F=(10**4,10**7,1000),us=(1000,20000,100),
              dmin=(10,2000,10))

display(y)
display(x) 
```
